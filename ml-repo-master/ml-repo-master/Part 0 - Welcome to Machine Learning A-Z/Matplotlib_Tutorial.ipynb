## Line Chart

```
#Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```
#Generates 1 D Array with 1000 evenly spaced array between 0-20
x = np.linspace(0, 20, 1000)
```

```
y = np.sin(x) + 0.2 * x
```

```
plt.plot(x, y)
plt.show()
```

```
plt.plot(x, y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('my first plot');
```

## Scatterplot

```
# Random data from Stadard Normal
X = np.random.randn(100, 2)
plt.scatter(X[:,0], X[:,1]); ## Specifing the X & Y dimension 
```

```
#rndom data with Standard Normal-- centered at 0.0
X = np.random.randn(200, 2)
#plt.scatter(X[:,0], X[:,1]); 
plt.scatter(X[:,0], X[:,1], c='r'); 
## Color argumenet 
```

```
#rndom data with Standard Normal-- centered at 0.0
X = np.random.randn(200, 2)
## Centered at 3
X[:50] += 3
##Generating label
Y = np.zeros(200)
##Setting forst 50 at 1
Y[:50] = 2
plt.scatter(X[:,0], X[:,1], c=Y); ## Color argumenet 
```
